```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Load image, grayscale, Otsu's threshold
image = cv2.imread('2.png')
print(image.shape)
image = cv2.resize(image,(600,800),interpolation = cv2.INTER_AREA)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 220, 255, cv2.THRESH_BINARY_INV)[1]
first =thresh.copy()
# Filter out large non-connecting objects
cnts= cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

cnts = cnts[0] if len(cnts) == 2 else cnts[1]
for c in cnts:
    area = cv2.contourArea(c)
    if area < 3:
        cv2.drawContours(thresh,[c],0,0,-1)
```

```
plt.figure(figsize=(20,20)),plt.imshow(thresh, cmap= 'gray')
```

```
new = thresh - first
n = np.where(new>0)
```

```
plt.figure(figsize=(20,20)),plt.imshow(new, cmap= 'gray')
```

```

# apply morphology
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (39,1))
img  = cv2.morphologyEx(new, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(20,20)),plt.imshow(img, cmap= 'gray')
```
